```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the data
models = ['GWO - 2 layers (Original)', 'GWO - 3 layers (Original)', 
          'GWO - 2 layers (PCA)', 'GWO - 3 layers (PCA)', 
          'GWO - 2 layers (RFECV)', 'GWO - 3 layers (RFECV)']

accuracies = [0.54793, 0.60231, 0.32359, 0.30455, 0.30048, 0.31203]

# Categorize models
categories = ['Original', 'Original', 'PCA', 'PCA', 'RFECV', 'RFECV']
layers = [2, 3, 2, 3, 2, 3]

# Create DataFrame
data = {
    'Model': models,
    'Accuracy': accuracies,
    'Category': categories,
    'Layers': layers
}
df = pd.DataFrame(data)

# Define custom colors
colors = ['#1f77b4', '#ff7f0e']  # Blue and Orange

# Create a plot
plt.figure(figsize=(12, 8))

# Plot grouped bar plot
sns.barplot(x='Category', y='Accuracy', hue='Layers', data=df, palette=colors, dodge=True, width=0.35)

# Add title and labels
plt.title('Accuracy Comparison by Data Preprocessing Method and Number of Layers')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.xlabel('Preprocessing Method')
plt.legend(title='Number of Layers', loc='lower right')

# Display the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Combine precision, recall, and f1-score into a single DataFrame
data = {
    'Model': np.repeat(models, len(activities)),
    'Activity': activities * len(models),
    'Precision': np.array(precision).flatten(),
    'Recall': np.array(recall).flatten(),
    'F1-Score': np.array(f1_score).flatten()
}
df = pd.DataFrame(data)

# Heatmap for each metric
for metric in ['Precision', 'Recall', 'F1-Score']:
    pivot_table = df.pivot(index='Activity', columns='Model', values=metric)
    plt.figure(figsize=(12, 6))
    sns.heatmap(pivot_table, annot=True, cmap='coolwarm', center=0)
    plt.title(f'{metric} Heatmap by Activity and Model')
    plt.show()
```
